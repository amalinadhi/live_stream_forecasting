```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
lb = 15
ub = 45
bs = np.linspace(lb, ub, 5)
bs
```

```python
def sigmoid(x):
    return 1/(1+np.exp(-x))

def sigmoid_green(x):
    demeaned_x = x - bs[-2]
    return sigmoid(demeaned_x)

def sigmoid_red(x):
    demeaned_x = x - bs[1]
    return sigmoid(-demeaned_x)
```

```python
x = np.linspace(-10, 80, 10001)
y_g = sigmoid_green(x)
y_r = sigmoid_red(x)
y_tot = y_g + y_r
```

```python
np.min(y_tot)
```

```python
plt.figure(figsize=(8, 3))

plt.plot(x, y_g, color='green', linewidth=1.5, linestyle='-.', label='UB sigmoid')
plt.plot(x, y_r, color='red', linewidth=1.5, linestyle='-.', label='LB sigmoid')
plt.plot(x, y_tot, color='black', linewidth=2.0, alpha=0.5, label='Summation')
plt.axvline(bs[1], color='grey', linestyle='--', linewidth=.5)
plt.axvline(bs[-2], color='grey', linestyle='--', linewidth=.5)
plt.axvline(bs[2], color='grey', linestyle='--', linewidth=.5)
plt.axvline(ub, color='grey', linestyle='--', linewidth=.5)
plt.axvline(lb, color='grey', linestyle='--', linewidth=.5)
plt.xlabel('x')
plt.ylabel('P(anomaly | x)')
plt.legend()
plt.show()
```

```python
mu = (ub+lb)/2.
p_mu = sigmoid_green(mu) + sigmoid_red(mu)

print(f'P(anomaly | x={mu:.2f}) = {p_mu:.4f}')

mu = lb
p_mu = sigmoid_green(mu) + sigmoid_red(mu)

print(f'P(anomaly | x={mu:.2f}) = {p_mu:.4f}')

mu = ub
p_mu = sigmoid_green(mu) + sigmoid_red(mu)

print(f'P(anomaly | x={mu:.2f}) = {p_mu:.4f}')
```
